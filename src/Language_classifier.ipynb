```python
import pandas as pd
import nltk
from nltk import word_tokenize, WordNetLemmatizer , pos_tag , word_tokenize
from nltk.corpus import wordnet
from nltk.corpus import stopwords
import string
nltk.download("stopwords")
```

```python
data_en= pd.read_csv('../data/sentiment_data.csv')
data_ar= pd.read_csv('../data/ar_reviews_100k.tsv', sep="\t")
```

```python
data_ar["language"]="arabic"
data_ar.drop(columns=["label"], inplace=True)
```

```python
data_en.columns
```

```python
data_en.drop(columns=['Unnamed: 0', 'Sentiment'], inplace=True)
```

```python
data_en.rename(columns={'Comment': 'text'}, inplace=True)
```

```python
data_en["language"]="english"
```

```python
data_ar=data_ar[:1000]
data_en=data_en[:1000]
```

```python
full_data=pd.concat([data_ar,data_en], axis=0)
```

```python
full_data
```

```python
full_data=full_data.sample(frac=1, random_state=42)
```

```python
full_data
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
y = full_data['language']
label = LabelEncoder()
y = label.fit_transform(y)
cv = TfidfVectorizer()
x = cv.fit_transform(full_data['text']).toarray()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x,y, train_size=0.8)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)

print(classification_report(y_test,y_pred))
```

```python
test=["شكرا"]

new_text=cv.transform(test).toarray()
prediction=model.predict(new_text)
lang=label.inverse_transform(prediction)
print(lang)
```
